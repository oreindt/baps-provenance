```julia
using JSON
```

```julia
map = JSON.parsefile("map_med1.json");
```

```julia
xs = []
ys = []
names = []
for c in map["cities"]
    push!(xs, c["x"])
    push!(ys, c["y"])
    push!(names, c["name"])
end
```

```julia
scatter(xs, ys, series_annotations=Plots.text.(names, 5))
```
